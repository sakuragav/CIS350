```python
import pandas as pd
used_cars = pd.read_csv("https://raw.githubusercontent.com/RBarroco/supervised-learning-regression/master/used_cars_data.csv")
used_cars.tail()
```

```python
df = pd.DataFrame(used_cars)
df = df.dropna()
df.tail()
```

```python
df['Brand'] = df['Name'].apply(lambda x: x.split(' ')[0])
df.head()
```

```python
def get_model(name):
  list = name.split(' ')
  list.pop(0)
  model = ' '.join(list)
  return model

df['Model'] = df['Name'].apply(lambda x: get_model(x))
df.head()
```
