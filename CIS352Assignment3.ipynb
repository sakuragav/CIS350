```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Sudz24/PlayStore/master/googleplaystore.csv")
df = pd.DataFrame(df)
df.head()
```

```python
def totalnan(df):
  df = pd.DataFrame(df)
  count_nan_in_df = df.isna().sum().sort_values(ascending=False)
  print(count_nan_in_df)

totalnan(df)
```

```python
# Write a Python code to remove the ‘+’ (plus) sign in the Installs feature. Make sure to include your code and output
def removeplus(num):
  return num.split('+')[0]
df['Installs'] = df['Installs'].apply(lambda x: removeplus(x))
df.head()
```

```python
# Use replace() function to write a Python code to remove the ‘,’ (comma) in the Installs feature.
def installsreplace(num):
  return num.replace(",", "")

df['Installs'] = df['Installs'].apply(lambda x: installsreplace(x))
df.head()
```

```python
# Use apply(), replace() and if functions to remove the letter ‘M’ in the Size feature.
def removeM(size):
    size = str(size)
    if size.split("M")[0] == "M":
      return size.replace("M", "")
    else:
      return size.replace("M", "")

df['Size'] = df['Size'].apply(lambda x: removeM(x))
df.head()
```
