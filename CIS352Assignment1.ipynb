```python
!pip install Pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[*]").getOrCreate()
from google.colab import files
uploaded = files.upload()
# Read the file
stock = spark.read.option("header", True).csv('apple_stock.csv')
stock.show()
stock.printSchema()
```

```python
# Compute the total daily Closing stock price and average of daily Low stock prices
from pyspark.sql.functions import col, sum, avg, max
stock.agg(sum("Close")).show()
stock.agg(avg("Low")).show()
```

```python
# Filter the data set based on opening stock price less than $500. Make your to include the code and output.
stock_filtered = stock.filter('Open < 500')
stock_filtered.show()
```
